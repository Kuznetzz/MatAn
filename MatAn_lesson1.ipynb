####  Задача 3 .
*****
    1.Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.
    2.Выполнить задание 1 на языке Python



```python
a = set(np.random.randint(5,20, size=15))
b = set(np.random.randint(10,25, size=15))
c = set(np.random.randint(15,30, size=15))
```

```python
letters = list(itertools.permutations('abc',2))
pairs = list(itertools.permutations([a,b,c],2))

print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")
print()

for i,j in enumerate (letters):
    print(f"Combination :{j}\n")
    
    print(f"Union: \n {pairs[i][0].union(pairs[i][1])}")
    
    print(f"Intersection: \n {pairs[i][0].intersection(pairs[i][1])}")
    
    print(f"Difference: \n {pairs[i][0].difference(pairs[i][1])}")
    
    print(f"Symmetric difference: \n {pairs[i][0].symmetric_difference(pairs[i][1])}")
    
    print(f"Cartesian product:")
    arr = list(itertools.product(pairs[i][0], pairs[i][1]))
    for k in range(len(arr))[::len(pairs[i][1])]:
        print(arr[k:k+len(pairs[i][1])])

    print(f"")
    print()
```

```python
from math import e

def root_factorial (n):
    return np.longdouble(n)/np.prod(np.power(np.arange(1,n+1,dtype=np.longdouble),np.longdouble(1/n)))

for i in range(1,11):
    print(root_factorial (i*10000000), "----", i*10000000 )
print(e,"math e")
```
